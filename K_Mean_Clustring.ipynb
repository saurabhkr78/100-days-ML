A mall wants  to understand its customer base to improve it's marketing strategy. the goal is to cluster customers based on their annual income and spending score. so the marketing team can tailor their campaigns to each cluster's preferences and behaviours.


```python

import numpy as np
import pandas as pd #group means table
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
data = {'AnnualIncome':np.random.randint(30000,100000,100),'spendingScore':np.random.randint(1,100,100)} #(min,max,noof)
df = pd.DataFrame(data)
df.head()
# print(df)
```

```python
plt.scatter(df['AnnualIncome'],df['spendingScore'])
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
#clustering customers
X=df.values
kmeans = KMeans(n_clusters=3,random_state=42)
```

```python
#fit dataframes
kmeans.fit(X)
```

```python
df['cluster']=kmeans.predict(X)
df.head()
```

```python
plt.scatter(df['AnnualIncome'],df['spendingScore'],c=df['cluster'],cmap='rainbow')
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
#testing
user_input={'AnnualIncome':[89000], 'spendingScore':[80]}
user_df=pd.DataFrame(user_input)
user_cluster=kmeans.predict(user_df)
print(f"The user belongs to cluster {user_cluster}")
```
